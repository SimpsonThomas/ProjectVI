```python
import numpy as np
from scipy.special import erf

def energy(config, J):
    # Compute the energy of a given configuration of spins
    L = config.shape[0]
    energy = 0
    for i in range(L):
        for j in range(L):
            energy += -J*config[i,j]*(config[(i+1)%L,j]+config[i,(j+1)%L])
    return energy

def magnetization(config):
    # Compute the magnetization of a given configuration of spins
    return np.sum(config)/config.size

def generate_data(L, J, T):
    # Generate a dataset of spin configurations
    data = []
    for i in range(1000):
        config = np.random.choice([-1, 1], size=(L, L))
        for t in range(100):
            # Perform a Monte Carlo step using the Metropolis algorithm
            x, y = np.random.randint(L, size=2)
            config[x, y] *= -1
            delta_E = -2*J*config[x,y]*(config[(x+1)%L,y]+config[x,(y+1)%L]+config[(x-1)%L,y]+config[x,(y-1)%L])
            if delta_E <= 0 or np.random.rand() < np.exp(-delta_E/T):
                continue
            config[x, y] *= -1
        data.append((magnetization(config), energy(config, J)))
    return np.array(data)
```

```python
from sklearn.metrics.pairwise import euclidean_distances

data = generate_data(L=10, J=1, T=0.5)
D = euclidean_distances(data)
```
